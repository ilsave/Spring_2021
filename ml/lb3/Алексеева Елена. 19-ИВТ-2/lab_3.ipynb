K-ближайших соседей (задаётся радиус для выбора ближайших объектов)

```python
from sklearn.metrics import accuracy_score, precision_recall_curve, average_precision_score, roc_curve, auc
from sklearn.metrics import plot_precision_recall_curve, plot_roc_curve

A = mushrooms_df.to_numpy()
Y = A[:,0]
X = A[:,1:]
# класс гриба запишем в переменную Y, остальные признаки - в X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1000, random_state=5)
# разделим датасет на обучающую и контрольную выборки
```

```python
from sklearn.neighbors import RadiusNeighborsClassifier
from sklearn.model_selection import GridSearchCV
RNN = RadiusNeighborsClassifier(outlier_label=1)
print(RNN.get_params())
# выбрали классификатор и выгрузили его стандартные параметры
```

```python
import numpy as np
parameters = {
    'radius': np.arange(0, 5, 0.25)
}
# выбираем параметры, изменения которых хотим исследовать
```

```python
RNN_search = GridSearchCV(RNN, parameters, n_jobs = 4)
RNN_search.fit(X_train, Y_train)
# Создаем GridSearchCV. Передаем наш классификатор и его параметры

best_radius = RNN_search.best_estimator_.radius
RNN = RadiusNeighborsClassifier(radius = best_radius)
RNN.fit(X_train, Y_train)

print(RNN_search.best_params_)
# выводим наилучший параметр
print("Accuracy of RNN model:", accuracy_score(Y_test, RNN.predict(X_test)))
print('Best accuracy:', RNN_search.best_score_)
```

Логический классификатор (бинарное решающее дерево)

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
# выбор классификатора
```

```python
print(DTC.get_params())
# вывод параметров по умолчанию
```

```python
parameters = {
    'max_depth': list(range(1,5)), # Максимальная глубина дерева
    'max_features': list(range(1,5)), # Количество признаков, которые следует учитывать при поиске лучшего разделения
}
# выбор параметров для исследования
```

```python
DTC_search = GridSearchCV(DTC, parameters, n_jobs=4)
# Создаем GridSearchCV. Передаем наш классификатор и его параметры
DTC_search.fit(X_train, Y_train)

best_depth = DTC_search.best_estimator_.max_depth # Находим лучшую глубину
best_features = DTC_search.best_estimator_.max_features # Находим лучшее количество признаков, которые следует учитывать при поиске

DTC = DecisionTreeClassifier(max_depth=best_depth, max_features=best_features)
DTC.fit(X_train, Y_train)

accuracy_score = np.mean(Y_test == DTC.predict(X_test))

print(DTC_search.best_params_)
# выводим наилучший параметр
print("Accuracy of DTC model:",  accuracy_score)
print("Best accuracy:", DTC_search.best_score_)
```

RP-кривая

```python
i = plot_precision_recall_curve(RNN, X_test, Y_test)
g = plot_precision_recall_curve(DTC, X_test, Y_test, ax = i.ax_)
```

ROC-кривая

```python
i = plot_roc_curve(RNN, X_test, Y_test)
g = plot_roc_curve(DTC, X_test, Y_test, ax = i.ax_)
```
