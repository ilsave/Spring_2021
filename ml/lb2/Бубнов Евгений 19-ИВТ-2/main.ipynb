```python
from sklearn.preprocessing import LabelEncoder
import scipy
import scipy.stats
import pandas as pd 
import numpy as np

label_encoder = LabelEncoder()


gender = np.asarray(astronauts['Space Walks'])
missions = np.asarray(label_encoder.fit_transform (astronauts['Missions']))


pirson = scipy.stats.pearsonr(gender, missions)
spirmen = scipy.stats.spearmanr(gender, missions)

print('Ð¿Ð¸Ñ€Ð¾ÑÐ¾Ð½ = ', pirson)

if (pirson[1] < 0.01):
    print("p-value < 0.01 Ð½ÐµÐ·Ð½Ð°Ñ‡Ð¸Ñ‚ÐµÐ»ÐµÐ½")
else:
    print("p-value > 0.01 - Ð·Ð½Ð°Ñ‡Ð¸Ñ‚ÐµÐ»ÐµÐ½")

print("Ð¡Ð¿Ð¸Ñ€Ð¼ÐµÐ½", spirmen)

if (spirmen[1] < 0.01):
    print("p-value < 0.01 Ð½ÐµÐ·Ð½Ð°Ñ‡Ð¸Ñ‚ÐµÐ»ÐµÐ½")
else:
    print("p-value > 0.01 - Ð·Ð½Ð°Ñ‡Ð¸Ñ‚ÐµÐ»ÐµÐ½")
```
