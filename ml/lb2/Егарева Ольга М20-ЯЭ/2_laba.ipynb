## 7. Постройте карту корреляций между признаками. Выведите также числовые значения признаков.

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(car_04['abtest'])
car_04['abtest'] = le.transform(car_04['abtest'])
le.fit(car_04['model'])
car_04['model'] = le.transform(car_04['model'])
le.fit(car_04['gearbox'])
car_04['gearbox'] = le.transform(car_04['gearbox'])
le.fit(car_04['fuelType'])
car_04['fuelType'] = le.transform(car_04['fuelType'])
le.fit(car_04['brand'])
car_04['brand'] = le.transform(car_04['brand'])
le.fit(car_04['notRepairedDamage'])
car_04['notRepairedDamage'] = le.transform(car_04['notRepairedDamage'])
le.fit(car_04['vehicleType'])
car_04['vehicleType'] = le.transform(car_04['vehicleType'])
```

## 8. Определите коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и типом кузова. Оцените значение p-value. Постройте плотность распределения признаков.

```python
from scipy import stats
corr_p, pvalue_p = stats.pearsonr(car_04['price'], car_04['vehicleType'])
print ("Коэффициент корреляции = ", corr_p)
print ("P-value = ", pvalue_p)
```

```python
corr_s, pvalue_s = stats.spearmanr(car_04['price'], car_04['vehicleType'])
print ("Коэффициент корреляции = ", corr_s)
print ("P-value = ", pvalue_s)
```
