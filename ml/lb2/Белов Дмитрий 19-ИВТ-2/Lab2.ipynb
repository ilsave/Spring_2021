# Задание 1
### Постройте графики количества осадков по годам в разных штатах.Сделайте выводы по построенным графикам.

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
sub1 = pd.DataFrame(rain_data[rain_data['SUBDIVISION'] == 'BIHAR']) # Загружаем данные в DataFrame и выбираем штат BIHAR
```

```python
x=sub1['YEAR'] #по оси x откладываем года
y=sub1['ANNUAL'] #по оси y количество осадков
plt.figure(figsize = (15,7)) #изменили размер фигуры(размер 15 на 7 дюймов)
plt.xticks(np.arange(1901,2016,2),rotation=90) #создание массива чисел(года), поворот надписи годов по оси на 90 градусов
plt.plot(x,y) #Вывод графика
plt.xlabel('Год') #Название колонки по оси х
plt.ylabel('Количество осадков') #название колонки по оси y
plt.title('Количество осадков по годам в штате BIHAR') #название графика
```

```python
sub2 = pd.DataFrame(rain_data[rain_data['SUBDIVISION'] == 'ORISSA']) # Загружаем данные в DataFrame и выбираем штат ORISSA
```

```python
x=sub2['YEAR'] #по оси x откладываем года
y=sub2['ANNUAL'] #по оси y количество осадков
plt.figure(figsize = (15,7)) #изменили размер фигуры(размер 15 на 7 дюймов)
plt.xticks(np.arange(1901,2016,2),rotation=90) #создание массива чисел(года), поворот надписи годов по оси на 90 градусов
plt.plot(x,y) #Вывод графика
plt.xlabel('Год') #Название колонки по оси х
plt.ylabel('Количество осадков') #название колонки по оси y
plt.title('Количество осадков по годам в штате ORISSA') #название графика
```

# Задание 9
### Вычислите корреляцию Пирсона и Спирмена между признаками NOV и MAR. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

#### Корреляцию Пирсона

```python
rain_data['NOV'].corr(rain_data['MAR'],method='pearson')
```

#### Корреляция Спирмена

```python
rain_data['NOV'].corr(rain_data['MAR'],method='spearman')
```

#### Вывод: Корреляция Спирмена больше корреляции Пирсона. 

```python
rain_data['NOV'].hist(bins=30) #bins -количство исполняемых ячеек
```

```python
rain_data['MAR'].hist(bins=30)
```

#### Вывод: Коэффициент Пирсона: p-value меньше 0.05, значит статистически незначима,то есть не существует связи между ноябрем и мартом.Коэффициент Спирмена: p-value менее 0.3, значит связь между признаками слабая.
